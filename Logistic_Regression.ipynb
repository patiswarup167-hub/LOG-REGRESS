1 : What is Simple Linear Regression (SLR)? Explain its purpose

ans-

    Simple Linear Regression (SLR) is a statistical technique that models the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.

    Purpose

    To predict the value of Y using X.

    To understand how Y changes when X changes.

    To find the strength and direction of the relationship.

    Example: Predicting a person’s weight based on their height.

 2: What are the key assumptions of Simple Linear Regression?

 ans-

    SLR relies on five main assumptions:

    Linearity
    The relationship between X and Y must be a straight line.

    Independence
    The observed data points should be independent of each other.

    Homoscedasticity
    The variance of errors must remain constant for all values of X.

    Normality of Errors
    The residuals (errors) should be normally distributed.

    No multicollinearity
    Not relevant for SLR (only one X), but important for multiple regression.

5: What is the method of least squares in linear regression?

ans-

    The Least Squares Method finds the best-fit line by minimizing the sum of squared errors (difference between actual and predicted values).
    It minimizes:
                 ∑(Yi​−Y^i​)2
        Where:


    Yi= actual value

    Y^i= predicted value

    This ensures the line has the smallest possible overall error.

7: Name and briefly describe three common evaluation metrics for regression
models.

ans-

      1. Mean Absolute Error (MAE)

    Average absolute difference between predicted and actual values.

    2. Mean Squared Error (MSE)

    Average of squared errors—penalizes large errors more.

    3. Root Mean Squared Error (RMSE)

    Square root of MSE; interpretable in the same units as Y.

8: What is the purpose of the R-squared metric in regression analysis?

ans-

    R-squared tells how much of the variation in Y is explained by X.

    Range: 0 to 1

    1 → Perfect fit

    0 → Model explains nothing

    Example:
    R² = 0.85 means 85% of the variation in Y is explained by X.

9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.

ans-



```python
import matplotlib.pyplot as plt
import numpy as np

# Re-using the data and model from the previous cell (cell_id: p2SNccOVbQP_)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
slope = 0.6
intercept = 2.2

# Create a scatter plot of the original data points
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual data points')

# Plot the regression line
# y_pred = slope * X + intercept
plt.plot(X, slope * X + intercept, color='red', label='Regression line')

# Add labels and title
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title('Simple Linear Regression Model')
plt.legend()
plt.grid(True)
plt.show()
```

10: How do you interpret the coefficients in a simple linear regression model?

ans-

    Slope (β₁) Interpretation

    For every 1-unit increase in X,
    Y changes by β₁ units.

    Example:
    β₁ = 3 → Y increases by 3 for every one-unit increase in X.

    ntercept (β₀) Interpretation

    The value of Y when X = 0.

    Example:
    β₀ = 5 → When X = 0, Y is predicted to be 5.
